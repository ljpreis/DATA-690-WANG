```python
#Setting column variable
columns = ["UNITID", "INSTNM", "STABBR", "REGION", "ADM_RATE", "TUITIONFEE_IN"]
```

```python
#Making a list of dataframes
df_list = [pd.read_csv(file, usecols=columns) for file in file_list]
    
```

```python
#Adding date to all of the dataframes
year =1995
for df in df_list:
    year = year+1
    df['Year'] = year
```

```python
#Making a big dataframe avaialble for exploration
big_df = pd.concat(df_list)
```

```python
#recasting the columns we're using as int and float
big_df["Year"] = big_df["Year"].astype(int)
big_df["TUITIONFEE_IN"] = big_df["TUITIONFEE_IN"].astype(float)
```

```python
#looking at the new large df
big_df
```

```python
#creating a new data frame for umbc
umbcdf=big_df.loc[big_df['INSTNM'] == 'University of Maryland-Baltimore County']


```

```python
umbcdf
```

```python
#Creating bar plot
fig, ax = plt.subplots(figsize=(12,8))
ax.bar(umbcdf["Year"],umbcdf["TUITIONFEE_IN"])
ax.set_title("UMBC Instate Tuition Overtime")
ax.set_xlabel("Year")
ax.set_ylabel("Tuition")

fig.show()
```

```python
#Creating line plot

fig, ax = plt.subplots(figsize=(12,8))
ax.plot(umbcdf["Year"],umbcdf["TUITIONFEE_IN"])
ax.set_title("UMBC Instate Tuition Overtime")
ax.set_xlabel("Year")
ax.set_ylabel("Tuition")

fig.show()
```

```python
#Calculating percent chagne

umbcdf["Tuition Percent Change"] = umbcdf['TUITIONFEE_IN'].pct_change()
```

```python
#Looking at the percent change colum
umbcdf
```

```python
#Creating bar plot
fig, ax = plt.subplots(figsize=(12,8))
ax.bar(umbcdf["Year"],umbcdf["Tuition Percent Change"])
ax.set_title("UMBC Instate Tuition Overtime")
ax.set_xlabel("Year")
ax.set_ylabel("Tuition")

fig.show()
```

```python
#Creating line plot

fig, ax = plt.subplots(figsize=(12,8))
ax.plot(umbcdf["Year"],umbcdf["Tuition Percent Change"])
ax.set_title("UMBC Instate Tuition Overtime")
ax.set_xlabel("Year")
ax.set_ylabel("Tuition")

fig.show()
```
